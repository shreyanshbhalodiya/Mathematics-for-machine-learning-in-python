## INVERSE OF MATRIX 

```python
A = np.random.randn(4,4)

A
```

```python
Ainv = np.linalg.inv(A)
Ainv
```

```python
AinvA = A @ Ainv
AinvA
```

```python
fig , ax = plt.subplots( 1 , 3 , figsize = (18, 16))
ax[0].imshow(A)
ax[0].set_title('A')

ax[1].imshow(Ainv)
ax[1].set_title("Inverse of A ")

ax[2].imshow(AinvA)
ax[2].set_title('AinvA = I (Identity matrix)')
plt.show()
```

## Third matrix looks like identity matrix

###### LATEX CODING TO WRITE A INVERSE SEE SECOND AND THIRD  BOX TITLE DIFFERENCE ABOVE AND BELOW

```python
fig , ax = plt.subplots( 1 , 3 , figsize = (18, 16))
ax[0].imshow(A)
ax[0].set_title('A')

ax[1].imshow(Ainv)
ax[1].set_title("A$^{-1}$")  ## latex coding to write A inverse 

ax[2].imshow(AinvA)
ax[2].set_title('A$^{-1}$A  = I')
plt.show()
```
